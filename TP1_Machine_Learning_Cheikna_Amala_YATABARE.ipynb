```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp

def analyze_train_test_split(X_train, X_test):
    """
    Fonction pour analyser la rÃ©partition entre les donnÃ©es d'entraÃ®nement et de test.
    Comprend la comparaison des distributions, l'analyse statistique, les matrices de corrÃ©lation,
    le test de Kolmogorov-Smirnov, et le calcul des diffÃ©rences relatives.
    """

    # 1. Comparaison des distributions
    print("\n--- Comparaison des distributions des variables ---\n")
    for column in X_train.columns:
        fig, axs = plt.subplots(1, 2, figsize=(14, 6))
        
        # Graphique de distribution
        sns.kdeplot(X_train[column], label="Train", color="blue", fill=True, ax=axs[0])
        sns.kdeplot(X_test[column], label="Test", color="red", fill=True, ax=axs[0])
        axs[0].set_title(f'Distribution de {column}')
        axs[0].legend()

        # Box plot
        data_to_plot = pd.DataFrame({'Train': X_train[column], 'Test': X_test[column]})
        sns.boxplot(data=data_to_plot, ax=axs[1])
        axs[1].set_title(f'Boxplot de {column}')
        axs[1].set_xticklabels(['Train', 'Test'])
        
        plt.tight_layout()
        plt.show()

    # 2. Analyse statistique descriptive
    print("\n--- Comparaison statistique descriptive ---\n")
    stats_comparison = pd.DataFrame({
        'Train_Mean': X_train.mean(),
        'Test_Mean': X_test.mean(),
        'Train_Std': X_train.std(),
        'Test_Std': X_test.std(),
        'Train_Min': X_train.min(),
        'Test_Min': X_test.min(),
        'Train_Max': X_train.max(),
        'Test_Max': X_test.max()
    })
    print(stats_comparison)

    # 3. Matrices de corrÃ©lation
    print("\n--- Matrices de corrÃ©lation ---\n")
    fig, axs = plt.subplots(1, 2, figsize=(16, 6))
    
    sns.heatmap(X_train.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, ax=axs[0])
    axs[0].set_title('CorrÃ©lations - Train')
    
    sns.heatmap(X_test.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, ax=axs[1])
    axs[1].set_title('CorrÃ©lations - Test')
    
    plt.tight_layout()
    plt.show()

    # 4. Test de Kolmogorov-Smirnov
    print("\n--- RÃ©sultats du test de Kolmogorov-Smirnov ---\n")
    ks_results = {}
    for column in X_train.columns:
        statistic, p_value = ks_2samp(X_train[column], X_test[column])
        ks_results[column] = {
            'KS_Statistic': statistic,
            'P_Value': p_value,
            'Distributions_Similaires': p_value > 0.05
        }
    ks_df = pd.DataFrame(ks_results).T
    print(ks_df)

    # 5. Calcul des diffÃ©rences relatives
    print("\n--- DiffÃ©rences relatives (%) entre Test et Train ---\n")
    rel_diff = pd.DataFrame({
        'Diff_Moyenne_%': ((X_test.mean() - X_train.mean()) / X_train.mean() * 100),
        'Diff_Std_%': ((X_test.std() - X_train.std()) / X_train.std() * 100)
    })
    print(rel_diff)

    return stats_comparison, ks_df, rel_diff

```

```python
class RegressionAnalyzer:
    def __init__(self, random_state: int = 42):
        """
        Initialize the regression analyzer with multiple models.
        
        Args:
            random_state (int): Random seed for reproducibility
        """
        self.random_state = random_state
        self.models = {
            'Ridge': Ridge(alpha=1.0),
            'Lasso': Lasso(alpha=0.1),
            'ElasticNet': ElasticNet(alpha=0.1, l1_ratio=0.5)
        }
        self.scaler = StandardScaler()
        self.results = {}

    def prepare_data(self, X: pd.DataFrame, y: pd.Series) -> Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
        """
        Prepare and split the data for modeling.
        
        Args:
            X (pd.DataFrame): Feature matrix
            y (pd.Series): Target variable
            
        Returns:
            Tuple containing training and test sets
        """
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.3, random_state=self.random_state
        )
        
        # Scale features
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)
        
        return X_train_scaled, X_test_scaled, y_train, y_test

    def train_and_evaluate(self, X_train: np.ndarray, X_test: np.ndarray, 
                          y_train: np.ndarray, y_test: np.ndarray) -> Dict:
        """
        Train models and compute predictions and metrics.
        
        Args:
            X_train, X_test, y_train, y_test: Training and test datasets
            
        Returns:
            Dictionary containing results for each model
        """
        for name, model in self.models.items():
            # Train model
            model.fit(X_train, y_train)
            
            # Make predictions
            y_pred_train = model.predict(X_train)
            y_pred_test = model.predict(X_test)
            
            # Compute metrics
            metrics = {
                'train': {
                    'mse': mean_squared_error(y_train, y_pred_train),
                    'rmse': np.sqrt(mean_squared_error(y_train, y_pred_train)),
                    'mae': mean_absolute_error(y_train, y_pred_train),
                    'r2': r2_score(y_train, y_pred_train)
                },
                'test': {
                    'mse': mean_squared_error(y_test, y_pred_test),
                    'rmse': np.sqrt(mean_squared_error(y_test, y_pred_test)),
                    'mae': mean_absolute_error(y_test, y_pred_test),
                    'r2': r2_score(y_test, y_pred_test)
                },
                'predictions': {
                    'train': {'actual': y_train, 'predicted': y_pred_train},
                    'test': {'actual': y_test, 'predicted': y_pred_test}
                }
            }
            
            # Add cross-validation scores
            cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
            metrics['cv_r2_mean'] = cv_scores.mean()
            metrics['cv_r2_std'] = cv_scores.std()
            
            self.results[name] = metrics

        return self.results

    def plot_predictions(self, title_fontsize: int = 12, figsize: Tuple[int, int] = (15, 5)):
        """
        Create enhanced scatter plots of predicted vs actual values.
        """
        for name, results in self.results.items():
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
            fig.suptitle(f'{name} Regression - Predictions vs Actual Values', 
                        fontsize=title_fontsize+2)
            
            # Training set
            self._plot_prediction_subplot(
                ax1, results['predictions']['train'],
                'Training Set', 'blue', name
            )
            
            # Test set
            self._plot_prediction_subplot(
                ax2, results['predictions']['test'],
                'Test Set', 'red', name
            )
            
            plt.tight_layout()
            plt.show()

    def _plot_prediction_subplot(self, ax, data: Dict, title: str, color: str, model_name: str):
        """Helper function for plotting prediction subplots."""
        actual, predicted = data['actual'], data['predicted']
        
        # Create scatter plot
        ax.scatter(actual, predicted, alpha=0.5, color=color, label='Predictions')
        
        # Add perfect prediction line
        min_val = min(actual.min(), predicted.min())
        max_val = max(actual.max(), predicted.max())
        ax.plot([min_val, max_val], [min_val, max_val], 'k--', 
                label='Perfect Predictions', alpha=0.8)
        
        # Add regression line
        z = np.polyfit(actual, predicted, 1)
        p = np.poly1d(z)
        ax.plot(actual, p(actual), color='green', alpha=0.8, 
                label=f'Trend Line (slope={z[0]:.2f})')
        
        # Customize plot
        ax.set_xlabel('Actual Values')
        ax.set_ylabel('Predicted Values')
        ax.set_title(f'{model_name} - {title}\n$R^2 = {r2_score(actual, predicted):.3f}$')
        ax.legend()
        ax.grid(True, alpha=0.3)

    def plot_residuals(self, figsize: Tuple[int, int] = (15, 5)):
        """
        Create enhanced residual plots with both boxplots and distributions.
        """
        for name, results in self.results.items():
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
            fig.suptitle(f'{name} Regression - Residual Analysis', fontsize=14)
            
            # Calculate residuals
            train_residuals = (results['predictions']['train']['actual'] - 
                             results['predictions']['train']['predicted'])
            test_residuals = (results['predictions']['test']['actual'] - 
                            results['predictions']['test']['predicted'])
            
            # Boxplot
            sns.boxplot(data=[train_residuals, test_residuals], 
                       ax=ax1, palette=['blue', 'red'])
            ax1.set_xticklabels(['Training', 'Test'])
            ax1.set_title('Residuals Distribution')
            ax1.set_ylabel('Residual Value')
            
            # Distribution plot
            sns.kdeplot(data=train_residuals, ax=ax2, color='blue', 
                       label='Training', fill=True, alpha=0.3)
            sns.kdeplot(data=test_residuals, ax=ax2, color='red', 
                       label='Test', fill=True, alpha=0.3)
            ax2.set_title('Residuals Density Distribution')
            ax2.set_xlabel('Residual Value')
            ax2.legend()
            
            plt.tight_layout()
            plt.show()

    def print_metrics(self):
        """
        Print comprehensive metrics for all models.
        """
        print("\n=== Model Performance Metrics ===")
        metrics_df = []
        
        for name, results in self.results.items():
            metrics = {
                'Model': name,
                'Train RÂ²': results['train']['r2'],
                'Test RÂ²': results['test']['r2'],
                'CV RÂ² (mean)': results['cv_r2_mean'],
                'CV RÂ² (std)': results['cv_r2_std'],
                'Train RMSE': results['train']['rmse'],
                'Test RMSE': results['test']['rmse'],
                'Train MAE': results['train']['mae'],
                'Test MAE': results['test']['mae']
                
            }
            metrics_df.append(metrics)
        
        df = pd.DataFrame(metrics_df).round(3)
        print(df.to_string(index=False))
        
        
        best_model = df.loc[df['Test RÂ²'].idxmax()]
        print(f"\nBest performing model: {best_model['Model']}")
        print(f"Test RÂ² score: {best_model['Test RÂ²']:.3f}")

def main(data: pd.DataFrame, target_col: str):
    """
    Main function to run the complete analysis.
    
    Args:
        data (pd.DataFrame): Input DataFrame
        target_col (str): Name of target column
    """
    
    data_cleaned = data.dropna(subset=[target_col])
    
    
    X = data_cleaned.drop(columns=[target_col])
    y = data_cleaned[target_col]
    
    
    analyzer = RegressionAnalyzer()
    X_train, X_test, y_train, y_test = analyzer.prepare_data(X, y)
    analyzer.train_and_evaluate(X_train, X_test, y_train, y_test)
    
   
    analyzer.plot_predictions()
    analyzer.plot_residuals()
    analyzer.print_metrics()
    
    return analyzer
```
